<h1> Primer ejemplo de series de tiempo en python <h1>

#### Se desea aprender los fundamentos de series de tiempo y Python
Serie de tiempo del promedio de lluvias, los datos estan en Lluvia.xlsx

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
#import seaborn as sn
```

```python
data = pd.read_excel('Lluvia.xlsx', index_col=0, parse_date=True) # El index es la primera columna y los datos son tiempos
#print(data)
print(type(data))
tm = pd.Series(data['lluvia'],data.index) # Convertir la data a serie de tiempo
#print(tm)
print(type(tm))
```

```python
plt.figure(figsize=(16,5))
plt.plot(tm)
plt.title('Promedio de lluvias mensuales')
plt.show
```
